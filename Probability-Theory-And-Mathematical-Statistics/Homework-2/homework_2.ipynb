## 🐍 **№1**

Подбрасываются четыре симметричные игральные кости (с шестью гранями). Обозначим через X минимальное значение выпавших очков (например, если выпало на первой кости 1, на второй – 2, на третьей – 6, на
четвертой – 1, тогда минимальное значение составляет 1). Найдите (аналитически) математическое ожидание E(X) и дисперсию Var(X). Разыграйте эксперимент на Python и найдите среднее значение такой величины. Постройте график зависимости среднего значения от числа экспериментов.

```python
def one_experiment() -> int:
  rolls = [np.random.randint(1, 7) for _ in range(4)]
  return min(rolls)

def simulate(number_of_experiments: int) -> tuple[list, list]:
  running_means = []
  running_vars = []

  total = 0
  total_squares = 0

  for index in range(1, number_of_experiments + 1):
    value = one_experiment()

    total += value
    total_squares += value ** 2

    mean = total / index
    var = (total_squares / index) - (mean ** 2)

    running_means.append(mean)
    running_vars.append(var)

  return (running_means, running_vars)

N = 100000

running_means, running_vars = simulate(N)

print(f'E(X) = {running_means[-1]}')
print(f'Var(X) = {running_vars[-1]}')
```

```python
figure = plt.figure(figsize = (9, 3))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(running_means)
axes.plot([running_means[-1] for _ in range(N)], color = 'red')
axes.set_xlabel('N', fontweight = 'bold')
axes.set_ylabel('E(X)', fontweight = 'bold')
axes.set_ylim(1.7, 1.77)

plt.show()
```

```python
figure = plt.figure(figsize = (9, 3))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(running_vars)
axes.plot([running_vars[-1] for _ in range(N)], color = 'red')
axes.set_xlabel('N', fontweight = 'bold')
axes.set_ylabel('Var(X)', fontweight = 'bold')
axes.set_ylim(0.850, 0.96)

plt.show()
```
