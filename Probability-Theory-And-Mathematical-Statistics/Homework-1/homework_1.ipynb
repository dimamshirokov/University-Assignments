## 🤝 **№1**

Два лица A и B условились встретиться в определенном месте между двумя и тремя часами дня. Лицо A ждет другого в течение 10 минут, после
чего уходит; лицо B ждет другого в течение 12 минут. Считая, что момент
прихода на встречу выбирается каждым «наудачу» в пределах указанного часа, найдите вероятность, что встреча состоится. Постройте график
зависимости относительной частоты pˆ(A)
def =
N(A)
N
события A от числа проведенных реализаций опыта N = 150 000 и приведете оценку вероятности
pˆ(A) = NA
N
для N = 150 000.
Ответ: 0, 332778. График зависимости относительной частоты pˆ(A)
def =
N(A)
N
события A от числа проведенных реализаций опыта N = 150 000 изображен на рисунке (оценка вероятности pˆ = 0.33268):

```python
N = 150000

a_list = np.random.uniform(0, 60, N)
b_list = np.random.uniform(0, 60, N)

condition = ((b_list >= a_list) & (b_list - a_list <= 10)) | ((a_list > b_list) & (a_list - b_list <= 12))

relative_frequency = np.cumsum(condition) / np.arange(1, N + 1)

probability = relative_frequency[-1]

figure = plt.figure(figsize = (9, 3))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(relative_frequency)
axes.plot([probability for _ in range(N)], color = 'red')
axes.set_xlabel('N', fontweight = 'bold')
axes.set_ylabel('Probability', fontweight = 'bold')
axes.set_ylim(0.3, 0.36)

plt.show()

print(f'\nProbability = {probability}')
```

## 👥 **№2**

часами. Каждый приходит и ждёт другого до истечения часа, но не более
15 минут. Найдите вероятности следующих событий: A – встреча состоялась; B – встреча не состоялась; C – Ивану не пришлось ждать Петра; D –
встреча состоялась после 11 ч. 30 мин; E – Иван опоздал на встречу; F –
встреча состоялась, когда до истечения часа оставалось меньше 5 минут.
Постройте график зависимости относительной частоты pˆ(A)
def =
N(A)
N
события A от числа проведенных реализаций опыта N = 150 000 и приведете
оценку вероятности pˆ(A) = NA
N
для N = 150 000.

```python
N = 150000

a_list = np.random.uniform(0, 60, N)
b_list = np.random.uniform(0, 60, N)

condition = np.abs(a_list - b_list) <= 15

A = condition
B = ~condition
C = condition & (a_list >= b_list)
D = condition & ((a_list + b_list) / 2 > 30)
E = a_list > b_list
F = condition & (np.maximum(a_list, b_list) > 55)

probability_A = np.mean(A)
probability_B = np.mean(B)
probability_C = np.mean(C)
probability_D = np.mean(D)
probability_E = np.mean(E)
probability_F = np.mean(F)

print(f'P(A) = {probability_A}')
print(f'P(B) = {probability_B}')
print(f'P(C) = {probability_C}')
print(f'P(D) = {probability_D}')
print(f'P(E) = {probability_E}')
print(f'P(F) = {probability_F}')
print()

relative_frequency = np.cumsum(condition) / np.arange(1, N + 1)

figure = plt.figure(figsize = (9, 3))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(relative_frequency)
axes.plot([probability_A for _ in range(N)], color = 'red')
axes.set_xlabel('N', fontweight = 'bold')
axes.set_ylabel('Probability', fontweight = 'bold')
axes.set_ylim(0.42, 0.45)

plt.show()
```

## ▽ **№5**

Отрезок [0; 1] случайным образом делится на 3 части. Найти вероятность
того, что из полученных отрезков можно сложить треугольник и этот треугольник будет остроугольным. Постройте график зависимости относительной частоты pˆ(A)
def =
N(A)
N
события A от числа проведенных реализаций опыта N = 150 000 и приведете оценку вероятности pˆ(A) = NA
N
для
N = 150 000. Найдите аналитическое решение для вероятности.

```python
N = 150000

x_list = np.random.uniform(0, 1, N)
y_list = np.array([np.random.uniform(x, 1) for x in x_list])

first_case = (x_list <= y_list) & (y_list >= 1 / 2) & (x_list <= 1 / 2) & (y_list <= x_list + 1 / 2)
second_case = (x_list) ** 2 + (y_list - 1) ** 2 >= (y_list - x_list) ** 2
third_case = (x_list) ** 2 + (y_list - x_list) ** 2 >= (y_list - 1) ** 2
fourth_case = (y_list - x_list) ** 2 + (y_list - 1) ** 2 >= (x_list) ** 2

condition = first_case & second_case & third_case & fourth_case

relative_frequency = np.cumsum(condition) / np.arange(1, N + 1)

probability = relative_frequency[-1]

figure = plt.figure(figsize = (9, 3))

axes = figure.add_axes([0, 0, 1, 1])
axes.plot(relative_frequency)
axes.plot([probability for _ in range(N)], color = 'red')
axes.set_xlabel('N', fontweight = 'bold')
axes.set_ylabel('Probability', fontweight = 'bold')
axes.set_ylim(0.055, 0.07)

plt.show()

print(f'\nProbability = {probability}')
```

## 〇 **№6**

В круге радиуса 1 наудачу выбирается отрезок. Найти вероятность того,
что длина этого отрезка будет меньше радиуса круга.

```python
N = 150000

R = 1

first_theta = np.random.uniform(0, 2 * np.pi, N)
r1 = np.sqrt(np.random.uniform(0, 1, N))
x1 = r1 * np.cos(first_theta)
y1 = r1 * np.sin(first_theta)

second_theta = np.random.uniform(0, 2 * np.pi, N)
r2 = np.sqrt(np.random.uniform(0, 1, N))
x2 = r2 * np.cos(second_theta)
y2 = r2 * np.sin(second_theta)

distances = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)

condition = distances < R

probability = np.mean(condition)

print(f'Probability = {probability}')
```

## ⬭ **№8**

В области, ограниченной эллипсом
(u/23)**2 + (v/6)**2, случайным образом
выбираются две точки. Пусть A и B – события, состоящие в том, что: A –
расстояние между выбранными точками меньше 9,2; B – координаты первой точки больше 0, а координаты второй точки меньше 0. Найдите приближенно, методом Монте-Карло: 1) вероятность P (A); 2) условную вероятность P (A|B). Указание: получите внутри заданного эллипса 100000
пар точек и, используя все эти пары точек, найдите ответы, округляя их
до одного знака после запятой.

```python
N = 100000

a = 23
b = 6

def generate_points_in_ellipse(N: int, a: int, b: int) -> tuple[np.array, np.array]:
  theta = np.random.uniform(0, 2 * np.pi, N)
  r = np.sqrt(np.random.uniform(0, 1, N))
  x = a * r * np.cos(theta)
  y = b * r * np.sin(theta)
  return (x, y)

x1, y1 = generate_points_in_ellipse(N, a, b)
x2, y2 = generate_points_in_ellipse(N, a, b)

condition_A = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2) < 9.2
condition_B = ((x1 > 0) & (y1 > 0)) & ((x2 < 0) & (y2 < 0))

probability_A = round(np.mean(condition_A), 2)
probability_A_B = round(np.mean(condition_A & condition_B) / np.mean(condition_B), 2)

print(f'P(A) = {probability_A}')
print(rf'P(A|B) = {probability_A_B}')
```

## ❒ **№9**

В кубе объема 1 случайным образом выбираются точки A, B и C. Пусть
R, S и T – события, состоящие в том, что: R – наименьший угол в треугольнике ABC меньше 26,7; S – наибольший угол в треугольнике ABC меньше 81,9; T – треугольник ABC остроугольный. Найдите приближенно, методом Монте-Карло: 1) условную вероятность P (R|T); 2) условную
вероятность P (S|T). Указание: получите 100000 остроугольных треугольников ABC и, используя все эти треугольники, найдите ответы, округляя
их до одного знака после запятой.

```python
N = 100000

def generate_points_in_cube(N: int) -> list[np.array, np.array, np.array]:
  x = np.random.uniform(0, 1, N)
  y = np.random.uniform(0, 1, N)
  z = np.random.uniform(0, 1, N)
  return (x, y, z)

def distance(x: tuple, y: tuple) -> np.array:
  x1, x2, x3 = x
  y1, y2, y3 = y
  return np.sqrt((x1 - y1) ** 2 + (x2 - y2) ** 2 + (x3 - y3) ** 2)

first_point = generate_points_in_cube(N)
second_point = generate_points_in_cube(N)
third_point = generate_points_in_cube(N)

a = distance(first_point, second_point)
b = distance(second_point, third_point)
c = distance(third_point, first_point)

cos_alpha = (a ** 2 + b ** 2 - c ** 2) / (2 * a * b)
cos_beta = (b ** 2 + c ** 2 - a ** 2) / (2 * b * c)
cos_gamma = (a ** 2 + c ** 2 - b ** 2) / (2 * a * c)

cos_alpha = np.clip(cos_alpha, -1, 1)
cos_beta = np.clip(cos_beta, -1, 1)
cos_gamma = np.clip(cos_gamma, -1, 1)

alpha = np.degrees(np.arccos(cos_alpha))
beta = np.degrees(np.arccos(cos_beta))
gamma = np.degrees(np.arccos(cos_gamma))

angles = np.array([alpha, beta, gamma])

condition_R = np.min(angles, axis = 0) < 26.7
condition_S = np.max(angles, axis = 0) < 81.9
condition_T = (alpha < 90) & (beta < 90) & (gamma < 90)

probability_R_T = round(np.mean(condition_R & condition_T) / np.mean(condition_T), 2)
probability_S_T = round(np.mean(condition_S & condition_T) / np.mean(condition_T), 2)

print(rf'P(R|T) = {probability_R_T}')
print(rf'P(S|T) = {probability_S_T}')
```
